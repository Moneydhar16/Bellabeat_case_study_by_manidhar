ROCCC stands for 

*  Reliable (R)

*  Original (O)

*  Comprehensive (C)

*  Current (C)

*  Cited (C)








Checking for ROCCC is a good measure for estimating the quality of data. 

Lets check ROCCC for the data in store ( Fitbit Fitness Tracker Data)


*   I'll be providing binary judgement using Low or High. Low for Disagreement and High for Agreement.

Current (Low) ( This dataset is collected 6 years back)

# 3.    Process Phase <a name="step3"></a>

```python
import pandas as pd
import numpy as np
import datetime as dt 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read the csv file using the pandas package 
df= pd.read_csv("/content/sample_data/dailyActivity_merged.csv")
df.head(5)
                                                                                                                     
```

```python
df.dtypes     # as seen ActivityDate is as object data type, should convert it to datetime
```

```python
df["ActivityDate"] = pd.to_datetime(df["ActivityDate"]) # converting the data type to datetime
```

```python
df.duplicated().sum() # checking for the duplicate entries 
#df.drop_duplicates(), incase if we find any duplicate entries 
```

```python
df.isnull().sum()    # no of null values column wise.
# df= df.interpolate() can be used to fill up the missing values of numerical dtype
```

```python
df.isna()
#df.fillna(0,inplace='true') incase error values are found replace with zero values
```

Checking for Data Integrity 

```python

df['Id'].nunique()         # unique function is used to get unique values and nunique gives the number of unique values 
# After running the above code we get the the no of ids has 33 which is a Data Integrity issue, as it was said the data was collected from 30 fitbit users.
```

```python
df['Day_of_the_week']= df['ActivityDate'].dt.day_name() # adding a day column for analysis
df['Day_of_the_week'].head(5)
```

```python
df.info() # To get a holistic information about the dataframe 
```

```python
new_cols= ['Id', 'ActivityDate', 'Day_of_the_week', 'TotalSteps', 'TotalDistance', 'TrackerDistance', 'LoggedActivitiesDistance', 'VeryActiveDistance', 'ModeratelyActiveDistance', 'LightActiveDistance', 'SedentaryActiveDistance', 'VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes', 'TotalExerciseMinutes', 'TotalExerciseHours', 'Calories']
df_new= df.reindex(columns=new_cols)

df_new.head(5)
```

```python
import numpy as np
from scipy import stats
# To remove outliers based on z scores
df_new.loc[(np.abs(stats.zscore(df_new['TotalSteps'])))>3,'outlier']='yes'
df_new.loc[(np.abs(stats.zscore(df_new['TotalSteps'])))<3,'outlier']='no'
df_new=df_new[df_new['outlier']=='no']
df_new
```

```python
# Creating a new columns that contains the total exercise hours
df_new['TotalExerciseMinutes']= df_new['VeryActiveMinutes']+df_new['FairlyActiveMinutes']+ df_new['LightlyActiveMinutes']+df_new['SedentaryMinutes']
df_new['TotalExerciseHours']=np.round(df_new['TotalExerciseMinutes']/60,2)
df_new.head()
```

# Act Phase <a name= step6></a>


 **1. Key Trends**

*   People are using the trackers mostly on the weekdays to track their fitness and Not on weekends which can be seen in the above visualization of daywise frequency plot

*   As can be seen in the above piechart, the tracker is mostly used to track the sedentary Minutes. From this, we can interpret that the fitness routine of the people is not that efficient or they are taking lots of breaks in between their fitness routine.

*   As can be seen in the categorical plot above, the median total steps taken by the users are the least on the weekend, especially Sunday.

*   The Total steps has a positive correlation with the calories burnt. Which means, Total steps is a good quantitative fitness measurement parameter.

*  Another key interpretation from the visualisations is that the median active percentage of users is less than 50 %. This data tells that the user activity effectiveness is very poor. 





**2. How could these trends apply to Bellabeat customers?**



*  Bellabeat Costumers are mostly women who uses smart devices to track their fitness data similar to the fitbit users data we analyzed. The trends obtained with the fitbit users might be equally valid for the Bellabeat consumers.

**3. How could these trends help influence Bellabeat marketing strategy?**



*   Bellabeat can levarge the fact the average user just completes just around 7637.9 steps. As recommended by CDC, an adult female has to aim at least 10,000 steps or 8km per day to benefit from general health, weight loss and fitness improvement( Source: [Medical News Today article](https://https://www.medicalnewstoday.com/articles/how-many-steps-should-you-take-a-day)). Bellabeat should make proactive advertisements emphasizing on this fact and educate women about their health. This would motivate women to buy Bellabeat products.

*   As we discussed earlier, sedentary minutes accounts for a large slice in the total exercise pie. Thus, Bellabeat products can add motivating notifications at workout intervals and also timely display milestones of calories burnt and Total steps.
